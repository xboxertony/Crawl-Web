```python
import quandl
import os
quandl.ApiConfig.api_key = os.environ.get('QUANDL_API_KEY') ## 取得金鑰來抓取資料
```

```python
quandl_code=['ISM/MAN_PMI','OPEC/ORB']
quandl_data=quandl.get(quandl_code,start_date='2008-12-31',collapse='monthly')
quandl_data
```

```python
from matplotlib import pyplot as plt

fig,ax1 = plt.subplots(figsize=(15,5))
line1, = ax1.plot(quandl_data.iloc[:,0],c='r')
ax1.set_ylabel('PMI')
ax2 = ax1.twinx()
line2, = ax2.plot(quandl_data.iloc[:,1],c='g')
ax2.set_ylabel('OPEC oil price')
plt.legend([line1,line2],['PMI','Oil Price'])
```
